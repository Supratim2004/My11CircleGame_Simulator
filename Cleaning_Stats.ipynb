```python
import os
import pandas as pd
import numpy as np
def convert_numpy(obj):
    if isinstance(obj, (np.int64, np.float64)):
        return obj.item()  # Convert to native Python types
    raise TypeError("Type not serializable")
```

```python
import os
folder="New Squads ESPN- improved"
subfolder=f"{folder}"
for fn in os.listdir(subfolder):
  if not os.path.isdir(os.path.join(subfolder, fn)):
    continue 
  if '.json' in fn:
    fn.removesuffix(".json") 
    os.rename(os.path.join(subfolder, fn),
            os.path.join(subfolder, fn.removesuffix(".json")))
  if '.json' not in fn:
    continue
```

```python
import json
Name_dict=dict()
for file in os.listdir(f"New Squads"):
    if file.endswith(".json"):
        name=file
        file=file.removesuffix(".json")
        print(file)
        name_dict=dict()
        for z in os.listdir(f"New Squads ESPN- improved/{file}"):
            if z.endswith(".xlsx"):
                l=z
                L=z.removesuffix(".xlsx").split("_")
                name_dict[L[1]]=[int(L[0]),f"New Squads ESPN- improved/{file}/{l}"]
        Name_dict[file]=name_dict
        with open(f"New Squads/{name}", 'r') as file:
            data = json.load(file)
        data1=data["player"]
        for player in data1:
            Player_name=player["name"]
            player["ESPN ID"]=name_dict[Player_name][0]
        with open(f"New Squads/{name}", "w") as json_file:
            json.dump(data, json_file, indent=4)
with open(f"Paths.json", "w") as json_file:
    json.dump(Name_dict, json_file, indent=4)
```

```python
from PlayerCard import *
import json
try:
    os.mkdir("Squads with Stats-improved/")
except FileExistsError:
    print("Directory already exists") 
with open(f"Paths.json", 'r') as file:
    data = json.load(file)
for i in data.keys():
    with open(f"New Squads/{i}.json", 'r') as file:
        data1 = json.load(file)
    z=data1["player"]
    for k in z:
        File_name="".join([f"New Squads ESPN- improved/{i}/","_".join([str(k["ESPN ID"]),k["name"]]),".xlsx"])
        Stats=pd.read_excel(File_name,sheet_name=None)
        Bat,Bowl,Field=Stats["Batting"],Stats["Bowling"],Stats["Fielding"]
        bat=Batting(Bat,"21-03-2024")
        bowl=Bowling(Bowl,"21-03-2024")
        field=Fielding(Field,"21-03-2024")
        batt=vars(bat)
        bowll=vars(bowl)
        fieldd=vars(field)
        del batt["data"],batt["Batting_Stats"],bowll["data"],bowll["Bowling_Stats"],fieldd["data"],fieldd["Fielding_Stats"]
        k["Batting Career Stats"],k["Bowling Career Stats"],k["Fielding Career Stats"]=batt,bowll,fieldd
    with open(f"Squads with Stats-improved/{i}.json", 'w') as json_file:
        print(data1)
        json.dump(data1, json_file,default=convert_numpy, indent=4)
```
