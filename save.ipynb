```python
import requests
import os
import time
from googlesearch import search
from bs4 import BeautifulSoup
import pandas as pd 
import csv
import re
import json
def get_career( player_id, match_format, data_type,file_name=None) :

        """Get Player career averages

        Arguments:
            file_name {string}: File name to save data
            match_format {int}: Match format (default is 11) (1-Test), (2-Odi) (3-T20I), (11-All International), (20-Youth Tests), (21-Youth ODI)
            data_type {string}: Data type (default is allround) (allround, batting, bowling, fielding)
        
        Return:
            Data in csv file
        """
        #self.match_format = match_format
        #self.data_type = data_type
        #self.file_name = file_name
# Perform the search and return the top results
        while(True):
            try:
                print(f"{player_id} {data_type}")
                if file_name is None:
                    file_name = f"{player_id}_{match_format}_{data_type}.csv"
                headers = {'user-agent': 'Chrome'}
                #https://stats.espncricinfo.com/ci/engine/player/{p}.html?class={match_format};template=results;type={data_type};view=match
                url=f"https://stats.espncricinfo.com/ci/engine/player/{player_id}.html?class={match_format};template=results;type={data_type};view=match"
                html_doc = requests.get(url, headers=headers)
                soup = BeautifulSoup(html_doc.text, 'html.parser')
                tables2 = soup.find_all("table")[2]
                table2_rows = tables2.find_all("tr")
                scores2 =[]
                for tr in table2_rows:
                    scores2.append(tr.text)
                tables = soup.find_all("table")[3]
                table_rows = tables.find_all("tr")
                scores =[]
                for tr in table_rows:
                    scores.append(tr.text)
                if scores2[0]!='\nNo records available to match this query\n':
                    #with pd.ExcelWriter('output.xlsx', engine='openpyxl') as writer:
                    with open(file_name, "w") as csv_file:
                        writer = csv.writer(csv_file, delimiter=',')
                        for row in scores:
                                writer.writerow(row.splitlines()) 
                    data=pd.read_csv(file_name)
                    os.remove(file_name)
                    return data
                else:
                    data=pd.DataFrame({"Availability":"No data available"},index=[0])
                    return data
                break
            except Exception as e:
                say()
                print(f"Attempt failed: {e}")
                time.sleep(2)
        
```
